There are a [variety of decoding strategies available](https://huggingface.co/docs/transformers/generation_strategies) in huggingface's `transformers` library. See the [huggingface blog](https://huggingface.co/blog/how-to-generate) for more information.
- `do_sample=True`: enables decoding strategies such as multinomial sampling, beam-search multinomial sampling, Top-K sampling and Top-p sampling.
- `max_new_tokens` = max_tokens: change max # of tokens
- `penatly_alpha = 0.6`:
- `top_k = 4`:
- `num_beams`: greater than 1 switches from greedy (default) to beam search.
- `num_return_sequences`: to return more than 1 sequence.



Different decoding methods:
```
#setup
prompt = "Hugging Face Company is"
inputs = tokenizer(prompt, return_tensors="pt")

#constrastive search
outputs = model.generate(**inputs, penalty_alpha=0.6, top_k=4, max_new_tokens=100)
tokenizer.batch_decode(outputs, skip_special_tokens=True)

#multinomial sampling
outputs = model.generate(**inputs, do_sample=True, max_new_tokens=100)
tokenizer.batch_decode(outputs, skip_special_tokens=True)

#beam search with multinomial sampling
outputs = model.generate(**inputs, num_beams=5, do_sample=True)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```
